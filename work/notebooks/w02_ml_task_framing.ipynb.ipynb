## 3. Success metric

*One metric you can defend. What number means 'good'?*

### Success metric: F1-score

I would use the F1-score as the main success metric because this is a binary classification task and both precision and recall matter. A false positive may waste review effort, while a false negative may cause a declining content item to be missed.

A higher F1-score means the model is better at balancing these two types of errors. I would compare the model against a simple baseline rather than choosing an arbitrary target score.
